I decided to switch to seaborn because it is much more well suited for dealing with pandas data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
manga = pd.read_csv("best-selling-manga.csv")
manga["year serialized"] = pd.to_numeric(manga["Serialized"].apply(lambda x: x.split("–")[0]))
manga = manga[manga.Demographic != "—"] # Removed unlabeled data
print(manga.dtypes)
```

```python
demographics = manga.groupby("Demographic").sum()
demographics
```

### Best Sellers

```python
manga.loc[0:3,["Manga series","Approximate sales in million(s)"]]
```

### Demographic Information

```python
ax = sns.barplot(data = manga, x = "Demographic",y="Approximate sales in million(s)")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
#plt.Show() 
```

```python
ax = sns.boxplot(data = manga, x = "Demographic",y="Approximate sales in million(s)")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

```python
ax = sns.barplot(data = manga, x = "Demographic",y="Average sales per volume in million(s)")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

```python
ax = sns.boxplot(data = manga, x = "Demographic",y="Average sales per volume in million(s)")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```
